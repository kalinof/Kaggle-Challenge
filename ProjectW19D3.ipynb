```python
###########
#LIBRARIES#     
###########
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import MinMaxScaler 
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.utils import resample
import pickle 
from sklearn.model_selection import cross_validate
import sys
import xgboost
import pickle #I'M GOING TO PICKLE THIS MODEL
```

```python
##########################
#SUMMARY OF DATA CLEANING#
##########################
sales=pd.read_csv('sales.csv')

# convert Unnamed to index
###########################
sales.rename(columns={'Unnamed: 0': 'index'}, inplace = True)
sales = sales.set_index('index')

# one hot encoding: state_holiday
##################
sales = pd.merge(left = sales,
                             right = pd.get_dummies(sales['state_holiday'],prefix='state_holiday'),
                             left_index=True,
                             right_index=True)


# date COLUMN:
##############
sales['date'] = pd.to_datetime(sales['date'])
sales['date'] = sales['date'].apply(lambda x: x.toordinal())

# one hot encoding: day_of_week
##################
sales = pd.merge(left = sales,
                             right = pd.get_dummies(sales['day_of_week'],prefix='weekday'),
                             left_index=True,
                             right_index=True)

#FEATURES AND LABELS SPLIT:
###########################
sales_final = sales.drop(['state_holiday','day_of_week'], axis=1, inplace=False)
features = sales_final.drop(columns=['sales','store_ID'],axis=1,inplace=False)
labels = sales_final['sales']
```

```python
###########
#  MODEL  #     
###########
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, random_state = 1)
xgb_reg = xgboost.XGBRegressor()
xgb_reg.fit(X_train, y_train)
xgb_reg.score(X_test,y_test)
print('test data accuracy was',xgb_reg.score(X_test,y_test))
print('train data accuracy was',xgb_reg.score(X_train,y_train))
```
